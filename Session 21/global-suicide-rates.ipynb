```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/kaggle/input/global-suicide-rates/30_merged_dataset_v00_final.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df.describe()
```

```python
df.info()
```

```python
df['Suicides number'] = df['Suicides number'].astype(int)
df['Suicides number'].dtype
```

```python
df['Adult Mortality'] = df['Adult Mortality'].astype(int)
df['Adult Mortality'].dtype
```

```python
df['Country'].nunique()
```

```python
df['Year'].nunique()
```

```python
df['Year'].unique()
```

```python
most_10_countries = df.groupby('Country')['Suicides number'].sum().sort_values(ascending=False).reset_index()
most_10_countries = most_10_countries.head(10)
most_10_countries
```

```python
plt.figure(figsize=(10, 6))
barplot = sns.barplot(x='Country', y='Suicides number', data=most_10_countries)
plt.xlabel('Country')
plt.ylabel('Total Suicides')
plt.xticks(rotation=45)
plt.title('Top 10 Countries by Total Suicides(2000 - 2011)')
```

```python
avg_5_countries = df.groupby('Country')['Suicides number'].mean().sort_values(ascending=False).reset_index()
avg_5_countries = avg_5_countries.head(5)
avg_5_countries
```

```python
plt.figure(figsize=(8, 8))
plt.pie(avg_5_countries['Suicides number'], labels=avg_5_countries['Country'], autopct='%1.1f%%', explode=[0.1,0,0,0,0])
plt.title('Average Suicides per Country (Top 5)')
```

```python
most_10_years = df.groupby('Year')['Suicides number'].sum().sort_values(ascending=False).reset_index()
most_10_years = most_10_years.head(10)
most_10_years
```

```python
plt.figure(figsize=(10, 6))
barplot = sns.barplot(x='Year', y='Suicides number', data=most_10_years)
plt.xlabel('Year')
plt.ylabel('Total Suicides')
plt.title('Top 10 Years by Total Suicides(2000 - 2011)')
```

```python
avg_5_years = df.groupby('Year')['Suicides number'].mean().sort_values(ascending=False).reset_index()
avg_5_years = avg_5_years.head(5)
avg_5_years
```

```python
plt.figure(figsize=(8, 8))
plt.pie(avg_5_years['Suicides number'], labels=avg_5_years['Year'], autopct='%1.1f%%', explode=[0.1,0,0,0,0])
plt.title('Average Suicides per Year (Top 5)')
```

```python
country_per_year = df.groupby(['Country','Year'])['Suicides number'].sum().sort_values(ascending=False).reset_index()
country_per_year = country_per_year.head(10)
country_per_year
```

```python
plt.figure(figsize=(12, 8))
barplot = sns.barplot(x='Year', y='Suicides number', hue='Country', data=country_per_year)
plt.xlabel('Country')
plt.ylabel('Total Suicides')
plt.title('Total Suicides per Country in one Year (Top 10)')
```
